```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("amazon_products_raw.csv")
df
```

```python
df.info()
```

```python
category_counts = df.groupby("main_category")["name"].nunique()

plt.bar(category_counts.index, category_counts.values)
plt.xlabel("Main Category")
plt.ylabel("Count of Product Names")
plt.title("Product Names Count by Main Category (n = 20)")
plt.xticks(range(len(category_counts)), [f"{i} ({count})" for i, count in zip(category_counts.index, category_counts.values)], rotation = 90)
plt.xticks(rotation = 90)
plt.show()
```
